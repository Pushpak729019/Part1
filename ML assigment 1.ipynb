```python
def remove_outliner(data1,col):
    Q1=data1[col].quantile(0.25)
    Q3=data1[col].quantile(0.75)
    IQR=Q3-Q1;
    lower=Q1-1.5*IQR
    upper=Q3+1.5*IQR
    data[col]=np.clip(data1[col], lower, upper)
    return data1

def treat_outliners_all(data1 , col_list): 
    for c in col_list: 
        data1 = remove_outliner(data , c) 
    return data1 
data = treat_outliners_all(data , data.iloc[: , 0::]) 
```

```python
travel_dist=[]
for pos in range(len(data['pickup_longitude'])):
    long1,lati1,long2,lati2=[data['pickup_longitude'][pos],data['pickup_latitude'][pos],
    data['dropoff_longitude'][pos],data['dropoff_latitude'][pos]]
    loc1=(lati1,long1)
    loc2=(lati2,long2)
    c=hs.haversine(loc1,loc2)
    travel_dist.append(c)
print(travel_dist)
data["dist_travel_km"]=travel_dist
data.head()
```
