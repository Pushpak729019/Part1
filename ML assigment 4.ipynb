```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return (x+3)**2

def df(x):
    return 2*(x+3)

def gradient_descent(initial_x,learning_rate,num_i):
    x = initial_x
    x_history = [x]
    for i in range(num_i):
        gradient=df(x)
        x=x-learning_rate*gradient
        x_history.append(x)
    return x,x_history
initial_x=0
learing_rate=0.1
num_i=50

x,x_history=gradient_descent(initial_x,learing_rate,num_i)
print("Local minimum: {:.2f}".format(x))
x_vals=np.linspace(-10,4,100)
plt.plot(x_vals,f(x_vals))
plt.plot(x_history,f(np.array(x_history)),'rx')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gradient Descent')
plt.show()
```

```python
cur_x = 3 # The algorithm starts at x=3 
rate = 0.01 # Learning rate 
precision = 0.000001 #This tells us when to stop the algorithm 
previous_step_size = 1 # 
max_iters = 10000 # maximum number of iterations 
iters = 0 #iteration counter 
df = lambda x: 2*(x+5) #Gradient of our function  
```

```python

while previous_step_size > precision and iters < max_iters: 
    prev_x = cur_x #Store current x value in prev_x 
    cur_x = cur_x - rate * df(prev_x) #Grad descent 
    previous_step_size = abs(cur_x - prev_x) #Change in x 
    iters = iters+1 #iteration count 
    print("Iteration",iters,"\nX value is",cur_x) #Print iterations 
    print("The local minimum occurs at", cur_x) 
```
